```python
df1=pd.DataFrame({"col1":list({"A","B","C"}),"col2":list({"A","B","C"})})
df2=pd.DataFrame({"col1":list({"A","B","C"}),"col2":list({"A","B","C"})})
```

```python
pd.concat([df1,df2])
```

```python
indexed_df2=df2.set_index("col1")
pd.merge(df1,indexed_df2,left_on="col1",right_index=True)
```

```python
pd.merge(df1,df2,on="col1",how="left")
```

```python
pd.merge(df1,df2,on="col1",how="right")
```

```python
pd.merge(df1,df2,on="col1",how="outer")
```
